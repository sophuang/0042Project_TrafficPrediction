## 1. Data Preprocessing

"SPATIAL has the simplest data preparation require- ment with each time series loaded into a single array of [Nsensors, Ntimepoints] for each of the three variables." 
需要每一个station ，都有两列【时间timestamp，volume】

```python
import pandas as pd

volume_df = pd.read_csv('volume_df.csv')
volume_df = pd.DataFrame(volume_df)
volume_df
```

```python
timeStamp = []

for day in range(23,30):
    for hour in range(0,24):
        string = "2023.01."+"% s" %day+" "+"% s" %hour+":00:00"
        timeStamp.append(string)

# Convert the list of strings to a DataFrame with a single column
timeStamp_df = pd.DataFrame({'timeStamp': timeStamp})

```

```python
volume_dfs = {}  # Create an empty dictionary to store the DataFrames

for row in range(len(volume_df)):
    volume = volume_df.iloc[row, 6:175]
    df = pd.concat([timeStamp_df, volume], axis=1)
    # Add the value of 'row' to the name of the DataFrame
    df.columns = ['timeStamp', 'volume']
    # Store the DataFrame in the dictionary
    volume_dfs[row] = df
```
